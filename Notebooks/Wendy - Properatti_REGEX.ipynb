```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import re
```

```python
df = pd.read_csv('properatti.csv')
```

```python
df.head()
```

```python
import unidecode

def quitar_caracteres(entrada):
    return str.lower(unidecode.unidecode(entrada))
```

```python
df['description'] = df['description'].astype('str').apply(quitar_caracteres)
df['title'] = df['title'].astype('str').apply(quitar_caracteres)
```

```python
regex_cochera = re.compile('cochera|garage')
df['regex_cochera'] = [regex_cochera.findall(x) for x in df['description']]

df['cochera'] = [1 if i != [] else 0 for i in df['regex_cochera']]
```

```python
regex_parilla = re.compile('parilla|asador|quincho')
df['regex_parilla'] = [regex_parilla.findall(x) for x in df['description']]

df['parilla'] = [1 if i != [] else 0 for i in df['regex_parilla']]
```

```python
regex_pileta= re.compile('pileta|piscina|jacuzzi')
df['regex_pileta'] = [regex_pileta.findall(x) for x in df['description']]

df['pileta'] = [1 if i != [] else 0 for i in df['regex_pileta']]
```

```python
regex_SUM = re.compile('sum|salon de usos multiples')
df['regex_SUM'] = [regex_SUM.findall(x) for x in df['description']]

df['SUM'] = [1 if i != [] else 0 for i in df['regex_SUM']]
```

```python
regex_gimnasio = re.compile('gimnasio')
df['regex_gimnasio'] = [regex_gimnasio.findall(x) for x in df['description']]

df['gimnasio'] = [1 if i != [] else 0 for i in df['regex_gimnasio']]
```

```python
regex_solarium = re.compile('solarium')
df['regex_solarium'] = [regex_solarium.findall(x) for x in df['description']]

df['solarium'] = [1 if i != [] else 0 for i in df['regex_solarium']]
```

```python
regex_lavadero = re.compile('laundry|lavadero')
df['regex_lavadero'] = [regex_lavadero.findall(x) for x in df['description']]

df['lavadero'] = [1 if i != [] else 0 for i in df['regex_lavadero']]
```

```python
regex_terraza = re.compile('terraza')
df['regex_terraza'] = [regex_terraza.findall(x) for x in df['description']]

df['terraza'] = [1 if i != [] else 0 for i in df['regex_terraza']]
```

```python
regex_patio = re.compile('patio')
df['regex_patio'] = [regex_patio.findall(x) for x in df['description']]

df['patio'] = [1 if i != [] else 0 for i in df['regex_patio']]
```

```python
regex_jardin = re.compile('jardin')
df['regex_jardin'] = [regex_jardin.findall(x) for x in df['description']]

df['jardin'] = [1 if i != [] else 0 for i in df['regex_jardin']]
```

```python
regex_balcon = re.compile('balcon')
df['regex_balcon'] = [regex_balcon.findall(x) for x in df['description']]

df['balcon'] = [1 if i != [] else 0 for i in df['regex_balcon']]
```

```python
df.drop(['regex_terraza','regex_balcon','regex_cochera','regex_parilla','regex_patio','regex_jardin','regex_pileta','regex_solarium','regex_gimnasio','regex_lavadero','regex_SUM'], axis=1, inplace=True)
```
